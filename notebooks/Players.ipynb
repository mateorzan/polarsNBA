```python
# Librerías
import polars as pl
import polars.selectors as cs
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.getcwd()

# Configurar visualización amplia 
pl.Config.set_tbl_cols(-1)
pl.Config.set_tbl_width_chars(300)
pl.Config.set_fmt_str_lengths(150)
pl.Config.set_tbl_rows(10)
```

### Cargamos el Dataset como LazyFrame

```python
# Install dependencies as needed:
# pip install kagglehub[polars-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "Players.csv"

# Load the latest version
lf = kagglehub.load_dataset(
  KaggleDatasetAdapter.POLARS,
  "eoinamoore/historical-nba-data-and-player-box-scores",
  file_path,
)

print("First 5 records:", lf.collect().head())
```

### Creamos el Dataframe para limpiarlo

```python
df_players = lf.collect()
df_players.head(2)
```

```python
# Columnas a eliminar que no aportan información relevante para el análisis
columnas_a_eliminar = ["birthDate"]

df_players = df_players.drop(columnas_a_eliminar)
```

```python
# Eliminamos todos los nulos ya que vemos que la mayoria de nulos son jugadores que no tienen datos de altura, peso, posición, etc. y no aportan información relevante para el análisis.

df_players = df_players.drop_nulls()
df_players.null_count()
```

### Outliers

```python
numeric_cols = df_players.select(cs.numeric()).columns


for col in numeric_cols:
    # Boxplot
    print(col)
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df_players[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```
