### Creamos el Dataframe para limpiarlo

```python
df_team_stats = lf.collect()
df_team_stats.head(2)
```

```python
# Eliminamos columnas que no aportan informaciГіn relevante para el anГЎlisis
columnas_a_eliminar = ["availableFlag", "teamName_right"]

df_team_stats = df_team_stats.drop(columnas_a_eliminar)
```

```python
df_team_stats.null_count()
```

```python
# Rellenar valores nulos con un valor especГӯfico
df_team_stats = df_team_stats.with_columns(
    pl.col("drebPct").fill_null(pl.col("drebPct").median()),
    pl.col("orebPct").fill_null(pl.col("orebPct").median()),
    pl.col("rebPct").fill_null(pl.col("rebPct").median())
)
```

```python
# Esto elimina cualquier fila donde CUALQUIERA de estas columnas sea null
df_team_stats = df_team_stats.filter(
    pl.col("teamCity").is_not_null(),
    pl.col("wl").is_not_null()
)
```
